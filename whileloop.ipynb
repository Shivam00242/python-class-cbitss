```python
# WAP to reverse the number
num=2345
reverse=0
while num>0:
  disit=num % 10
  num=num // 10
  print(disit)
  # reverse of num 2345
  reverse=reverse*10+disit
print(reverse)
```

```python
# WAP TO find number is palindrome or not
num=int(input("enter any number:"))
temp= num
reverse=0
while num >0:
  disit=num% 10
  num=num//10
  reverse=reverse*10+disit
print(num)
print(reverse)
if temp == reverse:
    print("number is palindrome")
else:
    print("number is not palindrome")

```
