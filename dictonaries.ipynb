```python
sales={
"c_name":["shivam","robin","aman","parth","aryan","shubham","saksham"],
"c_id":['101','102','103','104','105','106','107'],
"product":["shirt","t-shirt","iphone 17","samsung s25 ultra","shirt","t-shirt","iphone 17"],
"price":[2500,1500,150000,170000,1200,1500,149999],
"state":["hp","pb","hr","hr","hp","pb","hp"],
"gender":["m","f","f","m","m","f","m"]
}

unique_state=[]
state_wise_sale=[]
for i in sales["state"]:
  if i not in unique_state:
    unique_state.append(i)
    state_wise_sale.append(0)

for i in range(len(sales["state"])):
  index=unique_state.index(sales["state"][i])
  state_wise_sale[index]+=sales["price"][i]

print(state_wise_sale)


```
