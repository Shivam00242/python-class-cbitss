```python
#  restaurant bill with for loop
item=['pizza','pasta','chips','coke','dal chawal']
price=[450,250,40,80,150]
qty=[1,1,2,2,2]
discount=[12,21,25,32,11]

total_bill=0
total_bill_after_discount=0
print("_________One8 Commune________")
for i in range (0,len(item)):
  item_total=price[i]*qty[i]
  item_discount=(item_total*discount[i])/100
  item_after_discount=item_total-item_discount

  print("_________________  ",
        "\nitem=",item[i],
        "\n price=",price[i],
        "\n qty=",qty[i],
        "\n discounted rate=",price[i]-item_discount,
        "\n --------------"
        "\n item total=",item_after_discount,
        "\n")
  total_bill=total_bill+item_total
  total_bill_after_discount=total_bill_after_discount+item_after_discount
  final_bill=total_bill+total_bill_after_discount

print("____________________",
        "\n total bill=",final_bill)



```
