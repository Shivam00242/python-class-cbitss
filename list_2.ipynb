```python
# item,price,qty,discount
item=['pizza','chips','coke','pasta','roti','chicken','dal chawal']
price=[250,20,40,150,7,250,80]
qty=[1,2,1,1,5,1,1]
discount=[22,21,12,22,25,19,36]
print('---------bill----------')

total_bill_before_discount = 0
total_discount_amount = 0

for i in range (len(item)):
  item_total_price = qty[i] * price[i]
  item_discount_amount = (item_total_price * discount[i]) / 100
  item_price_after_discount = item_total_price - item_discount_amount

  print("item=",item[i],
        "\nprice=",price[i],
        "\nqty=",qty[i],
        "\ndiscount =",discount[i],"%",
        "\nitem total before discount =", item_total_price,
        "\ndiscount amount =", item_discount_amount,
        "\n------------")
  print("total of item", item[i], "is =", item_price_after_discount)
  print("________________")

  total_bill_before_discount += item_total_price
  total_discount_amount += item_discount_amount

grand_total = total_bill_before_discount - total_discount_amount

print("\nGrand Total before discount =", total_bill_before_discount)
print("Total Discount Applied =", total_discount_amount)
print("Grand Total after discount =", grand_total)
```
